```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import scale
%matplotlib inline
#%matplotlib notebook
plt.rcParams["figure.figsize"] = (8,6)
import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")
```

```python
df = pd.read_csv("World Happiness 2015-2019.csv")
df.head()
```

Bu veri seti, 2015'ten 2019'a kadar olan yllar arasnda dnya apndaki lkelerin mutluluk seviyelerini ieriyor. erik olarak u bilgileri barndryor:

Overall rank: lkenin o ylki mutluluk sralamasn gsterir.

Score: Mutluluk skoru. Bu skor, aada sralanan faktrlerin bir kombinasyonu olarak hesaplanmtr.

Social Support: Sosyal destek dzeyi.

Freedom to Choices: Bireylerin kendi hayatlar zerindeki zgrlk derecesi.

Generosity: Cmertlik dzeyi.

Perceptions of Corruption: Yolsuzluk algs.

```python
# Analizde kuLlanlacak feature'lar belirten bir degisken olusturalm

features = ["GDP","Social_Support","Health_life_expectancy","Freedom_to_choices", "Generosity", "Perceptions_of_corruption"]
```

```python
#veri tiplerini float olarak duzenleyelim

df[features] = df[features].astype("float")
```

```python
#overaL_rank (lkenin genel sralamas ) degerlerini integer yapalm

df["Overall rank"] = df["Overall rank"].astype("int")
```

```python
# Country degerlerini categoric data yapalm

df["Country"] = df["Country"].astype("category")
```

```python
#data setindeki region kategorisi null degerler ierdii ve kullanlmayaca iin drelim
df= df.drop(columns=['Region'])
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Perceptions_of_corruption'])
```

```python
# tekrar kontrol edelim ve null deger yok

df.isnull().sum()
```

```python
# overall_rank deerlerini baz aLarok LkeLeri 5 kategoriye ayralm

category = pd.qcut(df["Overall rank"], 5, labels=[ "Very Top Rank","Top rank","Middle Rank","Low Rank","Very Low Rank"])
```

```python
# df iinde bulunan lkelerin ait oLduklar kategoriLeri class degiskenine atayalm

df["class"] = category
```

## Scaling

PCA uygularken zelliklerin leklendirilmesi, daha dengeli ve doru sonular elde etmek iin nemli bir admdr..

Farkl leklerdeki feature; temel bir bileenin yksek varyansl deikene baml olmasna yol aacaktr. Bu istenmeyen bir durumdur. Bu aamada **ortak leklendirme yaparak veriyi normalletirmek PCA iin elzemdir**

```python
# DeikenLeri scale yapalm

df_scaled = scale(df[features ])
```

```python
# Box-plot ile lkelerin feature'lara gre dalm

df2 = pd.DataFrame( df_scaled, columns=features)
df2["Country"] = pd.Series ( df["Country"] , index=df.index)
df3 = pd.melt(df2, id_vars="Country", value_vars=df2[features])
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.boxplot(y= "variable", x="value", data=df3, palette="Set2")
plt.show()
```

```python
# means ve std kontrol

df2[features].describe().loc[["count", "mean", "std"]].round()
```

```python
# feature'lar aras ilikiyi scatter plot ile inceleme

sns.set(style="whitegrid")
sns.pairplot(df2[features], kind = "reg", diag_kind="kde")
plt.show()
```

```python
# degiskenler arasndaki korelasyon

plt.figure(figsize=(6,4))
sns.set(style="whitegrid")
sns.heatmap(df2[features].corr(method="pearson"), vmin=-.1, vmax=1, annot=True, cmap="RdYlGn")
plt.show()
```

## PCA hyperparameter

```python
# pip install prince
```

```python
import prince
pca = prince.PCA(n_components=6, n_iter=10, rescale_with_mean=False,
rescale_with_std=False, copy=True, check_input=True,
engine= "sklearn" , random_state=42)

# PCA modeli iin hiperporometrelere deerleri girelim:
# PCA 56 feature iin yaplacak. Zaten 6 bileen vard, hepsini dikkate alyoruz.
```

```python
# PCA modelini pca variable a atayalm

pca = pca.fit(df2[features])
```

- PCA hibir feature  dlamadna dahil eder
- ok sayda feature overfitting e gtrebilir

**NOT-2**:  Bu aamada kmeleme asndan bizi ilgilendiren **PCA'nn hesaplad temel bileenlerden kann incelenen olayn ounu akladdr**. Yani, **hesaplanan temel bileenlerden ka anlaml kmeleme iin yeterlidir?** Bunun iin Temel bileenlere ait elde edilen z vektr deerleri (eigen values) ya da varyans grafiinin bklme noktasnn solunda kalan temel bileenler seilir

```python
#Modelin eigeanvalues deerleri

eigeanvalues = pca.eigenvalues_
print(eigeanvalues)
```

```python
# Elbow metodu ile optimum temel bileen says seimi

ev = pd.DataFrame()
ev["pca"] = range(1,7)
ev["eigenvalue"]=pd.DataFrame(pca.eigenvalues_)
plt.figure(figsize=(8,6))
sns.lineplot(x="pca", y = "eigenvalue", marker = "o", data= ev)
plt.ylabel("Eigenvalue", fontsize=16)
plt.xlabel("Temel Bileenler", fontsize=16)
plt.show()
```

stteki grafikte varyansn ounun ilk 3 bileenle (component) aklanabilecei grlyor.

```python
# AkLanan Varyans Oranlar (kmlatif)
# Hangi temel bileen incelenen olay ne kadar aklyor? sorusunun cevab

explained_variance=pca.cumulative_percentage_of_variance_
print(explained_variance)
```

yani 1. bileen %45, 2.bileen %(69-45) orannda aklyor 

```python
# avo: Aklanan varyans oran (kmlatif)
# Temel Kural: Ka temel bileen ile en az %80 varyans aklanyor?

avo = pd.DataFrame()
avo["pca"] = range(1,7)
avo["vari"] = pd.DataFrame(pca.cumulative_percentage_of_variance_)

plt.figure(figsize=(8,6))
graph = sns.barplot(x="pca", y ="vari", data=avo)

for p in graph.patches :
    graph.annotate('{:.2f}'.format(p.get_height()),(p.get_x()+0.2,p.get_height()), ha="center", va="bottom", color="black")

plt.ylabel("Aklanan Varyans Oran (%)", fontsize=18)
plt.xlabel("Temel Bileenler", fontsize=18)
plt.show()  
```

stteki grafkten de anlalaca zere 1 . Temel Bileen toplam varyansn %46'sini, 1. ve 2. Temel Bileen toplamda  %69'n ve 1-2-3. Temel Bileen toplamlar da  % 80'ini aklamaktadr. 
Yani  birden hikayenin (varyansn) %80'ini aklyor. 
**Pratik bir kural olarak (rule of thumb) %80 ve zeri aklanan varyans eik deer olarak kabul edilebilir**

**NEML**: Elbette u soru akla gelebilir 6 feature'  3'e indirgeme iin onca dntrme ilemini yapmaya deer mi? 
Cevap: 6 feature  iin bunca zahmete demeyebilir. Ama her aracn veri rettii gnmz dnyasnda 1OO'lerce feature'dan oluan veri setleri sz konusuysa gerekten buna deer.

```python
# Column (stun) korelasyonu 

correlations = pca.column_correlations
correlations
```

```python
# Deikenlerin Temel Bileen 1 ve 2 iindeki Pozisyonlar Grafii 

scatter = pca.column_correlations.reset_index()

plt.figure(figsize=(10,6))

ax = sns.scatterplot(x=0, y=1, data=scatter)
ax.set(ylim=(-1,1), xlim=(-1,0))

def label_point(x, y, val, ax):
    a = pd.concat({"x": x, "y": y, "val": pd.Series(val)}, axis=1)  # scatter.index'i Series olarak kullan
    for i, point in a.iterrows():
        ax.text(point["x"]+.02, point["y"], str(point["val"]))

label_point(scatter[0], scatter[1], scatter.index, plt.gca())
plt.axvline(-0.5, ls="--")
plt.axhline(0, ls="--")
plt.title("Deikenlerin Temel Bileen 1 ve 2 iindeki Pozisyonlar", fontsize=18)
plt.xlabel("Temel Bileen 1 (49%)", fontsize=18)
plt.ylabel("Temel Bileen 2 (24%)", fontsize=18)
plt.show()
```

```python
# Deikenlerin Temel Bileen 1 ve 2 iindeki Pozisyonlar Grafii
scatter = pca.column_correlations.reset_index()


plt.figure(figsize=(10,6))
ax = sns.scatterplot(x=0, y=1, data=scatter)
ax.set(ylim=(-1,1), xlim=(-1,0))
def label_point(x, y, val,lab, ax):
    a = pd.concat({"x": x, "y": y, "val": pd.Series(val),"a":lab}, axis=1)  # scatter.index'i Series olarak kullan
    for i, point in a.iterrows():
        ax.text(point["x"]+.02, point["y"], f'{str(point["val"])} {str(point["a"])}' )
label_point(scatter[0], scatter[1], scatter.index, scatter.variable,plt.gca())
plt.axvline(-0.5, ls="--")
plt.axhline(0, ls="--")
plt.title("Deikenlerin Temel Bileen 1 ve 2 iindeki Pozisyonlar", fontsize=18)
plt.xlabel("Temel Bileen 1 (49%)", fontsize=18)
plt.ylabel("Temel Bileen 2 (24%)", fontsize=18)
plt.show()
```

**NOT**: stteki grafik, hangi feature'larn birbirine yakn olduunu (feature'lar aras gruplamalar) gstermesi bakmndan olduka faydal bir yaklam. lk iki temel bileene dayal  deiken grup olduu grlyor: i) GSYH-GDP, sosyal destek-Social Support ve salkl yaam beklentisi-Healty life exp, ii) (st grafikte 3.0 ve 5.0 noktas)yaam seimleri yapma zgrl-Freedom  to make ve yolsuzluk algs-Perception ve iii) cmertlik-Generosity (grafikte 4.0 noktas).

## Cluster Prediction

```python
# lkelerin ait olduu kmeleri tahmin edelim

# 1. ve 2. biLeen iftine gre hmeLeme tahmini

x_pca = pca.transform(df2[features])
x_pca.head()
```

```python
# Ulkelerin 5 kumeye (classes) ayrlm grntsn elde etme 
row_coords = pca.row_coordinates(df2[features])

plt.figure(figsize=(6,6))
ax = sns.scatterplot(
    x=row_coords.iloc[:, 0],  
    y=row_coords.iloc[:, 1],  
    hue=df["class"]           
)

ax.set_title("PCA Row Coordinates")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Farkl bileen iftlerine gre kmeleme tahminleri

pd.DataFrame.iteritems = pd.DataFrame.items
import plotly.express as px
from plotly.offline import plot
components = pca.fit_transform(df2[features])
labels = {str(i): f"PC {i+1} ({var:.1f}%)"
          for i, var in enumerate(pca.cumulative_percentage_of_variance_)}
fig = px.scatter_matrix(components, labels=labels, dimensions=range(4),
                        color=df["class"])
fig.update_traces(diagonal_visible=False)
plot(fig)
```
